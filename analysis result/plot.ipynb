```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for MTTR ratios
mttr_data = {
    "Language": ["surgujia", "kudukh", "halbi", "Biaganai", "Gondi", "sadri", "chhattisgarhi"],
    "1-gram": [0.0998, 0.1021, 0.0918, 0.0950, 0.0852, 0.0888, 0.0931],
    "2-gram": [0.3916, 0.4874, 0.4128, 0.3934, 0.4001, 0.3906, 0.4136],
    "3-gram": [0.6405, 0.7730, 0.6849, 0.6344, 0.6598, 0.6377, 0.6857],
    "4-gram": [0.7778, 0.8873, 0.8179, 0.7554, 0.7882, 0.7729, 0.8197],
    "5-gram": [0.8573, 0.9341, 0.8848, 0.8250, 0.8545, 0.8494, 0.8880],
    "6-gram": [0.9028, 0.9599, 0.9237, 0.8718, 0.8960, 0.8975, 0.9263],
    "7-gram": [0.9305, 0.9749, 0.9486, 0.9042, 0.9248, 0.9266, 0.9498]
}

# Creating DataFrame
mttr_df = pd.DataFrame(mttr_data)

# Plotting n-gram vs. MTTR ratios for all dialects
plt.figure(figsize=(10, 6))

n_grams = ["1-gram", "2-gram", "3-gram", "4-gram", "5-gram", "6-gram", "7-gram"]

for index, row in mttr_df.iterrows():
    plt.plot(n_grams, row[1:], marker='o', label=row["Language"])

plt.xlabel("N-gram")
plt.ylabel("MTTR Ratio")
plt.title("N-gram vs. MTTR Ratio for Different Dialects")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
